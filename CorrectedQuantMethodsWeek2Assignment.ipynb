# 1. (5 points) Write a code to implement the elementary random walk Xn
# by using a random number generator to choose the direction for each
# step. Run your code once for N = 10, 000 and p = 0.5 (symmetric
# random walk).

```python
# Python code for 1-D random walk.
import random
import numpy as np
import matplotlib.pyplot as plt
 
def genRandomWalk(p, N):

    # statically defining the starting position
    start = 0 
    positions = [start]

    # creating the random points
    Yi = np.random.binomial(size = N, n = 1, p = p)

    # changing all zeros to -1
    for j in range(len(Yi)):
        if(Yi[j] == 0):
            Yi[j] = -1

    # keep the starting position 0
    Yi = np.insert(Yi,0, 0, axis = 0)
    # calculate random walk
    Xi = np.cumsum(Yi)
    
    return Xi

# generating random walk
Xi = genRandomWalk(0.5, 10000)
```

# {Xn|0 ≤ n ≤ N }

```python
 
# plotting down the graph of the random walk in 1D
plt.plot(Xi)
plt.show()
```

# 5 points) Modify the code to implement the rescaled random walk Wn
# and plot one sample path for p = 0.5 and N = 10, 000

```python
def BrownianMotion(p, N):
    Wi = genRandomWalk(p, N)
    Nj = 0
    for j in range(len(Wi)):
        
        # calculating the divisor for rescaling
        Nj = j + 1
        sqrtNj = Nj**(1/2)
        
        # rescaling
        Wi[j] /= sqrtNj
    
    return Wi

Wn = BrownianMotion(0.5, 10000)
# plotting down the graph of the brownian motion in 1D
plt.plot(Wn)
plt.show()
```

# 6. (5 points) Plot now 1, 000 paths of Wn for N = 10, 000

```python
for j in range(1000):
    Wn = BrownianMotion(0.5, 10000)
    # plotting down the graph of the brownian motion in 1D
    plt.plot(Wn)

plt.show()
```
