# of this process. We set σ = 0.2, Y0 = 0, and λ = 2 per year.

1. (20 points) Write a code in Matlab, R or Python to simulate this process
over a time period of 10 years and plot the simulated paths.

```python
# Python code for 1-D random walk.
import random
import numpy as np
import matplotlib.pyplot as plt
 
def genPoissonRandomWalk(sigma,Y0, lamda, noOfYears, subtractTrend=False):

    # creating the poisson residuals
    ei = np.random.poisson(lamda, noOfYears)

    Yi = [Y0]
    
    # calculate the random walk
    for j in range(noOfYears):
        sm = 0
        for k in range(j):
            if(subtractTrend):
                sm += ei[k] - lamda
            else:
                sm += ei[k]
        
      
        Yi.append(Yi[0] + sigma*sm )
    Yi.pop(0)
    return Yi

# generating random walk
Yi = genPoissonRandomWalk(0.2, 0, 2, 10)
```

```python
# plotting down the graph of the poisson random walk
plt.plot(Yi)
plt.show()
```

```python
for j in range(100):
    Yt = genPoissonRandomWalk(0.2, 0, 2, 10)
    # plotting down the graph of the brownian motion in 1D
    plt.plot(Yt)

plt.show()
```

Visually the paths do not look like a martingale as there is an upward trend in them. If they were martinagales there wouldn't be any trend either upward or downward. As the all paths are rising they are discrete time submartinagales with respect to themselves

What does the plot suggest

```python
for j in range(100):
    Yt = genPoissonRandomWalk(0.2, 0, 1, 10, subtractTrend=True)
    # plotting down the graph of the brownian motion in 1D
    plt.plot(Yt)
```

The plot has no trend either upward or downward. By visual inspection we can say that the random walk is now a martingale

5. (10 points) Could the above models be used to model a stock price?
What would their major flaws be? Discuss.

The above model can be used to model stock prices where we would assume that current stock price depends on previous days stock price and a poisson distributed residual.  
The major flaw of this model:

1) This model assumes that stock prices have memorylessness property which may not be true in reality
2) The above model can not model volatility/variance properly as it assumes no trend in the prices. But in reality the market has bullish(upward) or bearish(downwards) trends
3) The assumption for poisson distribution for residuals generally does not hold true which will make our model perform worse
